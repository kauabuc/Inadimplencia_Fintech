# 3. Análise Univariada 

Nessa parte irei explorar cada variável, afim de entender possíveis causas que colaboram com a inadimplência, além também de descobrir possíveis padrões.

## 3.1 Idade

```python
# Tatal de Clientes por grupo

grupos = [18, 28, 38, 48, 75]

df["idade_agrupada"] = pd.cut(df["Idade"], grupos)

freq_table = pd.crosstab(df['idade_agrupada'], columns='count')
freq_table
```

```python
# Frequencia relativa em relação aos grupos

freq_table / freq_table.sum()
```

```python
plt.title("Distribuição de Idade")

sns.histplot(df["Idade"], kde=True);
```

```python
# Medidas resumo da variável

df["Idade"].describe()
```

```python
plt.title("Box-Plot de Idade")

sns.boxplot(data=df["Idade"]);
```

Com isso obtive os seguintes insigths: 
- A média de idade dos clientes é de 35 anos e mediana de 33 anos.
- 75% dos clientes apresentaram ter até 42 anos de idade.
- A maior concentração de clientes está na faixa etária de 28-38

## 3.2 Sexo

```python
plt.title("Proporção de Sexo")

valores = df["Sexo"].value_counts()

plt.pie(valores, labels=["Mulheres", "Homens"], colors=["rosybrown", "skyblue"], autopct='%1.1f%%')
plt.show()
```

Podemos concluir que temos mais mulheres (69%) em nosso conjunto, porém os homens apresentam ter uma maior taxa de inadimplência. Irei explorar isso ao longo do projeto.

# 4 Análise Bivariada

Aqui procuro entender as relações das váriaveis uma com as outras, e também buscando padrões que possam explicar que sejam explicativos em relação a inadimplencia. Para isso, será necessário usar as técnicas de Correlação de Pearson e Análise IV

## 4.1 Correlação de Pearson

```python
df_correlacao = df[["Idade", "Duracao_Emprestimo", "Valor_Emprestimo", "Inadimplencia"]]
df_correlacao = df_correlacao.corr()


sns.heatmap(data=df_correlacao, annot=True, cmap='Greens');
```

```python
# Criação de gráficos de dispersão em relação a Inadimplencia
sns.pairplot(df,hue='Inadimplencia');
```

Em relação a correlação linear, apenas a duração do empréstimo junto ao valor do empréstimo apresentam uma correlação significativa sendo 0,62,  o que faz sentido no caso, quando aumenta o valor do empréstimo o cliente tende a pedir mais meses para poder pagar.

## 4.2 Análise IV

```python
sc.iv(df_original, "Inadimplencia")
```

```python
y = "Inadimplencia"

bins = sc.woebin(df_original, y);
```

```python
sc.woebin_plot(bins);
```

Obtemos que: 
- A variável que mais conta com poder de separação é a de Saldo de Conta Corrente sendo 0,73
- Variáveis que também contam com um poder de separação médio, no caso Saldo investido, valor emprestimo e duração do emprestimo, também podem indicar padrões na inadimplência.
- Como as outras variáveis apresentam um poder de separação fraco ou muito fraco, poderiamos descarta-lás caso tivesse um alto custo para a empresa resgatas essas variáveis.

# 5. Conclusão

Partindo disso podemos tirar algumas conclusões relacionadas a Inadimplencia:
1. Os clientes que não possuem conta em outro banco, apresentam maiores chances de realizarem o pagamento do crédito
2. Já os clientes que tem pouco saldo ou um saldo moderado em algum outro banco, apresentam uma maior chance a não realizarem o pagamento do crédito
3. Os clientes que possuem saldo investido em outro banco, sendo um saldo baixo ou moderado, também apresentam uma chance maior de não concluirem o pagamento do empréstimo
4. Já em relação aos empréstimos, é necessário ficar de olho nos clientes que pedem empréstimo com uma duração maior que 25 meses, onde os mesmos são propensos a inadimplentes.
5. Com o valor, clientes que pedem um valor acima de 4000, também apresentam maior chance de inadimplência.
6. Clientes que pedem emprésitmo de até 10 meses, são mais propensos a pagarem o crédito pedido.
7. Em relação ao saldo em outro banco, no caso de clientes que não tem saldo investido e clientes sem conta serem mais propensos a pagar pois em teoria, o cliente não teria como simplesmente migrar para outro banco, fazendo com que ele seja obrigado a pagar a dívida para continuar utilizando o serviço da empresa.

Com essas conclusões, agora pode ser feita a criação de medidas e planos de ações a partir dos dados existentes de modo que diminua a atual inadimplencia dos clientes.
